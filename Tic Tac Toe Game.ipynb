```python
# TIC TAC TOE GAME 

a=1
b=2
c=3
d=4
e=5
f=6
g=7
h=8
i=9
value1 = 0
value2 = 0
marker = False



def game_on():
    print("Hello, Welcome to Tic Tae Toe!")


    
def display():
    
    print("   |   |   ")
    print(a," |",b,"| ",c)
    print("   |   |   ")
    print("-----------")
    print("   |   |   ")
    print(d," |",e,"| ",f)
    print("   |   |   ")
    print("-----------")
    print("   |   |   ")
    print(g," |",h,"| ",i)
    print("   |   |   ")


    
def player_choice():
    choice = False
    
    while choice == False:
        print("Player 1: Do you want to be X or O?")
        z=input("Player 1 Enter your choice:")
        
        if z == "X" or z == "O":
            if z == "X":
                print("Player 1 is X")
                choice = True
            else:
                print("Player 1 is O")
                choice = True
        else:
            print("Sorry! Select either X or O")
            
        return z
           

        
def player_X():
    acceptable_range = range(1,10)
    within_range = False
    
    while within_range == False:
        
        position = int(input("Select a position from 1 to 9:"))
        
        if position in acceptable_range:
            within_range = True
        else:
            print("Sorry! Enter a number with in range 1 to 9")
        
    return position



def player_O():
    acceptable_range = range(1,10)
    within_range = False
    
    while within_range == False:
        
        position = int(input("Select a position from 1 to 9:"))
        
        if position in acceptable_range:
            within_range = True
        else:
            print("Sorry! Enter a number with in range 1 to 9")
        
    return position



def update_game_X():
    global value1,a,b,c,d,e,f,g,h,i
    if value1 == 1:
        a="X"
    elif value1 == 2:
        b="X"
    elif value1 == 3:
        c="X"
    elif value1 == 4:
        d="X"
    elif value1 == 5:
        e="X"
    elif value1 == 6:
        f="X"
    elif value1 == 7:
        g="X"
    elif value1 == 8:
        h="X"
    else:
        i="X"
        

        
def update_game_O():
    global value2,a,b,c,d,e,f,g,h,i
    if value2 == 1:
        a="O"
    elif value2 == 2:
        b="O"
    elif value2 == 3:
        c="O"
    elif value2 == 4:
        d="O"
    elif value2 == 5:
        e="O"
    elif value2 == 6:
        f="O"
    elif value2 == 7:
        g="O"
    elif value2 == 8:
        h="O"
    else:
        i="O"



def check():
    global marker
    if a=="X" and b=="X" and c=="X":
        print("Player 1 has won!")
        marker = True
    elif d=="X" and e=="X" and f=="X":
        print("Player 1 has won!")
        marker = True
    elif g=="X" and h=="X" and i=="X":
        print("Player 1 has won!")
        marker = True
    elif a=="X" and d=="X" and g=="X":
        print("Player 1 has won!")
        marker = True
    elif b=="X" and e=="X" and h=="X":
        print("Player 1 has won!")
        marker = True
    elif c=="X" and f=="X" and i=="X":
        print("Player 1 has won!")
        marker = True
    elif a=="X" and e=="X" and i=="X":
        print("Player 1 has won!")
        marker = True
    elif c=="X" and e=="X" and g=="X":
        print("Player 1 has won!")
        marker = True
    elif a=="O" and b=="O" and c=="O":
        print("Player 2 has won!")
        marker = True
    elif d=="O" and e=="O" and f=="O":
        print("Player 2 has won!")
        marker = True
    elif g=="O" and h=="O" and i=="O":
        print("Player 2 has won!")
        marker = True
    elif a=="O" and d=="O" and g=="O":
        print("Player 2 has won!")
        marker = True
    elif b=="O" and e=="O" and h=="O":
        print("Player 2 has won!")
        marker = True
    elif c=="O" and f=="O" and i=="O":
        print("Player 2 has won!")
        marker = True
    elif a=="O" and e=="O" and i=="O":
        print("Player 2 has won!")
        marker = True
    elif c=="O" and e=="O" and g=="O":
        print("Player 2 has won!")
        marker = True
    else:
        pass
        

        
def game():
    while True:
        game_on()
        display()
        user_choice=player_choice()
        
        while True:
            global a,b,c,d,e,f,g,h,i,value1,value2,marker
            if user_choice == "X":
                value1 = player_X()
                update_game_X()
                display()
                check()
                if marker == True:
                    break
                else:
                    pass
                print("Player 2 : It is your chance!")
                value2 = player_O()
                update_game_O()
                display()
                check()
                if marker == True:
                    break
                else:
                    pass
            
            else:
                value2 = player_O()
                update_game_O()
                display()
                check()
                if marker == True:
                    break
                else:
                    pass
                print("Player 2: It is your chance!")
                value1 = player_X()
                update_game_X()
                display()
                check()
                if marker == True:
                    break
                else:
                    pass
            
        continue_game = input("Do you want to continue the game (Y/N)")

        if continue_game == "Y":
            continue
        else:
            break

game()
```
